# Photometry

Let's use the WDP to obtain the time series. It is structured in 2 steps:

1. Get catalog;
2. Assemble light curve.

```python
# Lets get the catalog of sources

from wdpipe.photometry.aperture_phot import assemble_lightcurve, get_catalog

light_curve = assemble_lightcurve(object_folder + "/final",
                                  cat, final_parameters_file, 
                                  aperture_factors={'r': 1.5, 'r_in': 2.0, 'r_out': 3.0})
cat, _ = get_catalog(ref_img, ref_pars.to_dict(), nsigma=5)
```

```python
# Let's assemble the lightcurve

time_series_file = "time_series_21mai03_30.csv"

if not exists(time_series_file):
    light_curve = assemble_lightcurve(object_folder + "/final",
                                      cat, final_parameters_file, 
                                      aperture_factors={'r': 1.5, 'r_in': 2.0, 'r_out': 3.0}
                                     )
    np.savetxt(time_series_file, light_curve, delimiter=",")
else:
    light_curve = np.loadtxt(time_series_file, delimiter=",")
```

```python
fig, ax = plt.subplots()

means_1 = np.nanmean(light_curve[:, 3::2], axis=1)
stds_1 = np.nanstd(light_curve[:, 3::2], axis=1, ddof=1)

ax.scatter(means_1, stds_1, s=15, alpha=.5)
ax.set_xlim(11, 19)
ax.set_ylim(0, 0.2)
ax.set_xlabel("$V$")
ax.set_ylabel(f"$\sigma_V$")
ax.yaxis.grid()
ax.set_title(f"Série instrumental")
fig.tight_layout()
```

# Stellar Cluster Differential Photometry

```python
from wdpipe.photometry.differential_phot import differential_photometry
ref_mags, ref_stars, diff_ts = differential_photometry(light_curve, n=5, viz=True)
```

```python
fig, [ax1, ax2] = plt.subplots(nrows=2)

y_b = 0.05


means = np.nanmean(diff_ts[:, 3:], axis=1)
stds = np.nanstd(diff_ts[:, 3:], axis=1, ddof=1)

means_1 = np.nanmean(light_curve[:, 3::2], axis=1)
stds_1 = np.nanstd(light_curve[:, 3::2], axis=1, ddof=1)

ax1.scatter(means_1, stds_1, s=15, alpha=.5)
ax1.set_xlim(11, 19)
ax1.set_ylim(0, y_b)
ax1.set_xlabel("$V$")
ax1.set_ylabel(f"$\sigma_V$")
ax1.yaxis.grid()
ax1.set_title(f"Série instrumental")


ax2.scatter(means, stds, s=15, alpha=.5)
ax2.set_xlim(-2, 6)
ax2.set_ylim(0, y_b)
ax2.set_xlabel("$V$")
ax2.set_ylabel(f"$\sigma_V$")
ax2.yaxis.grid()
ax2.set_title(f"Série diferencial")
fig.tight_layout()
```
